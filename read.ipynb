```python
import joblib

# ลองโหลดไฟล์ด้วย joblib
try:
    model = joblib.load('export-jsps001-1h-arima.pkl')
    print("ไฟล์ถูกโหลดเรียบร้อยด้วย joblib!")
except Exception as e:
    print(f"เกิดข้อผิดพลาด: {e}")
```

```python
# ตรวจสอบประเภทของวัตถุที่โหลดมา
print(f"ประเภทของวัตถุ: {type(model)}")

# ตรวจสอบเมธอดและแอตทริบิวต์ที่ใช้งานได้
if hasattr(model, '__dict__'):
    print("แอตทริบิวต์ของโมเดล:")
    for key, value in model.__dict__.items():
        print(f"{key}: {value}")
else:
    print("วัตถุนี้ไม่มีแอตทริบิวต์ __dict__")

# ตรวจสอบว่าโมเดลมีเมธอด predict หรือไม่
if hasattr(model, 'predict'):
    print("โมเดลนี้มีเมธอด predict")
else:
    print("โมเดลนี้ไม่มีเมธอด predict")
```
